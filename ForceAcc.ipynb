```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

car_data = pd.read_csv('Automobile.csv')
car_data["force"] = car_data["weight"].mul(car_data["acceleration"])
X = car_data[["weight", "force"]]/100
y = car_data["acceleration"]
```

```python
sns.lmplot(x="weight", y="force", data=car_data)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
score = r2_score(predictions, y_test)
MAE = mean_absolute_error(predictions, y_test)
print(f'{score*100} %')
```
